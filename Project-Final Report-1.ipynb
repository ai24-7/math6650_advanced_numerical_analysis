### 1. Introduction

The paper focuses on Hermite interpolation and extends and generalizes an iterative method introduced by Cirillo et al. in 2018, originally applied to the Floater–Hormann family of interpolants. This new approach yields an effective barycentric rational trigonometric Hermite interpolant at general ordered nodes. The foundation of this work is based on basis functions from Berrut's 1988 trigonometric interpolant.

A key advancement in this research is the analytical calculation of the differentiation matrix, facilitating easier computational construction of the interpolant. The paper provides various examples and a thorough numerical study examining the convergence at equidistant and conformally mapped nodes. It outlines an iterative scheme for constructing Hermite interpolants from general interpolation bases, focusing specifically on Berrut’s trigonometric interpolant. The resulting interpolant demonstrates potential applications in fields such as algorithm development, image processing, and network reliability analysis.

### 2. Proof Checking


#### 2.1 Lemma 1
This lemma establishes the foundation for constructing Hermite interpolants. It involves basis functions $b_i(x)$ satisfying the Lagrange property at nodes. These functions, combined with a smooth function $d_i(x) $ that vanishes at $x_i$ and satisfies $d'_i(x_i) = 1$, lead to the creation of Hermite interpolants satisfying the Hermite conditions. This forms the backbone of the iterative construction process.


![lemma](lemma.png)

The proof that is provided by the author is mentioned below.

![lemma_proof](lemma_proof.png)

I provide a proof by induction.

Base Case:

$$b_{i,0}^{(k)} = \frac{1}{0!} d_i(x)^0 b_i(x)^1 = b_i(x)$$
Given the fact that $b_i(x)$ satisfy Lagrange property, i.e. $b_i(x_i) = \delta_{i,j}$, and kronecker delta is zero when $i=j$ and one when $i\neq j $  \
Therefore base case holds.

Inductive Hypothesis: \
Assume that for $k<j-1$ :
$$b_{i,j-1}^{(k)}(x_p) = 0$$
$$b_{i,j-1}^{(j-1)}(x_p) = \delta_{i,p}$$

Let $ c_i(x) = d_i(x) b_i(x)$ \
and \
$$ b_{i,j}(x) = \frac{1}{j!} d_i(x) b_i(x) d_i(x)^{j-1} b_i(x)^j$$
Therefore,
$$ b_{i,j}(x) = \frac{1}{j!} c_i(x) d_i(x)^{j-1} b_i(x)^j $$
We also have 
$$ b_{i,j-1}(x) = \frac{1}{(j-1)!} d_i(x)^{j-1} b_i(x)^j $$ \
Then we can rewrite $b_{i,j}$ as : \
$$ b_{i,j}(x) = \frac{1}{j} \frac{1}{(j-1)!} c_i(x) d_i(x)^{j-1} b_i(x)^j = \frac{1}{j} c_i(x) b_{i,j-1}(x) $$
Now, we can use generalized Leibniz Rule of product for two functions f and g:
$$\frac{d^k}{dx^k}(f(x) g(x)) = \sum_{s=0}^{k} {k \choose s} f^{(s)}(x) g^{(k-s)}(x) $$
To compute k-th derivative of basis:
$$ b_{i,j}^{(k)}(x) = \frac{1}{j} \sum_{s=0}^k {k \choose s} c_i^{(k-s)}(x) b_{i,j-1}^{(s)}(x) $$

Now from inductive hypothesis, for $s<j-1$ we have: \
$$ b_{i,j-1}^{(s)}(x_p) = 0 $$
Therefore: \
$$ b_{i,j}^{(k)}(x_p) = \frac{1}{j} \sum_{s=j-1}^{k-1} {k \choose s} c_i^{(k-s)}(x_p) b_{i,j-1}^{(s)}(x_p)$$

Now for $k<j$ summation is empty. Therefore: \
$$b_{i,j}^{(k)}(x_p) = 0 $$

for $k=j$ we have: \
$$ b_{i,j}^{(j)}(x_p) = \frac{1}{j} {j \choose j-1} c_i'(x_p) b_{i,j-1}^{(j-1)}(x_p) $$

From combination formula,
$$ {j \choose j-1} = j$$ \
From inductive hypothesis,
$$ b_{i,j-1}^{(j-1)}(x_p) = \delta_{i,p} $$
Then we have:
$$b_{i,j}^{(j)}(x_p) = c_i'(x_p) \delta_{i,p}(x_p) $$

Now let's focus on $c_i'(x)$ \
we have $c_i(x) = d_i(x)b_i(x)$ \
so \
$$ c_i'(x_p) = d_i'(x_p) b_i(x_p) + d_i(x_p) b_i'(xp)$$
In the beginning, we assumed that $d(x)$ is a smooth function with three properties, \
one of them is that $d_i'(xp)=1$ . we also know from induction hypothesis that $b_i'(x_p) = 0$. \
Finally, due to the result of base case, we know that $b_i(x_p)=1$ when $i\neq p$ . \
Therefore : \
$$ c_i'(x_p) = 1$$

And : \
$$b_{i,j}^{(j)}(x_p) = \delta_{i,p}(x_p) $$

The proof is done!

#### 2.2 Theorem 1

Theorem 1 proves the effectiveness of the iterative approach for creating Hermite interpolants. It asserts that the function $r_m$, constructed using basis functions $b_{i,j}$ and coefficients $g_{i,j}$ , satisfies the Hermite conditions. This theorem gives us the theoretical foundation of the barycentric rational trigonometric Hermite interpolant proposed in this paper.

![theorem1](theorem1.png)


The author provides this proof: 
![theorem1_proof](theorem_proof.png)

My proof is : \
If we take k_th derivative of $r_m$ we have :
$$ r_m^{(k)}(x_l) = \sum_{i=0}^n \sum_{j=0}^m b_{i,j}^{(k)}(x)\bigg|_{x=x_l} g_{i,j} $$
Using the result of lemma 1, for $j>k$ we have 0, so
$$ r_m^{(k)}(x_l) = \sum_{i=0}^n \sum_{j=0}^k b_{i,j}^{(k)}(x)\bigg|_{x=x_l} g_{i,j} $$
$$ r_m^{(k)}(x_l) = \sum_{i=0}^n b_{i,k}^{(k)}(x)\bigg|_{x=x_l} g_{i,k} +\sum_{i=0}^n \sum_{j=0}^{k-1} b_{i,j}^{(k)}(x)\bigg|_{x=x_l} g_{i,j} $$
From the proof of lemma 1, we know that $b_{i,k}^{(k)}(x_l)= \delta_{i,l} $ and all terms are zeros where $i \neq l$. We also know that $\delta_{l,l}=1$. Therefore, the only terms that get out of summation is $g_{l,k}$.
$$ r_m^{(k)}(x_l) = g_{l,k} +\sum_{i=0}^n \sum_{j=0}^{k-1} b_{i,j}^{(k)}(x)\bigg|_{x=x_l} g_{i,j} $$
$$ r_m^{(k)}(x_l) = g_{l,k} + r_{k-1}^{(k)}(x_l) $$
$$ r_m^{(k)}(x_l) = f_{l,k}$$
Therefore, it satisfies Hermite conditions and proof is done!

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

# Cst function used in Berrut's trigonometric interpolant
def cst(phi, n):
    if n % 2 == 0:
        x = 1. / np.tan(phi)
    else:
        x = 1. / np.sin(phi)
    return x
    
# Calculate up to third derivatives of first periodic test function
def f(x):
    return np.exp(2 * np.sin(x) + np.cos(x))

def Df(x):
    return np.exp(2 * np.sin(x) + np.cos(x)) * (2 * np.cos(x) - np.sin(x))

def D2f(x):
    return np.exp(np.cos(x) + 2 * np.sin(x)) * (4 * np.cos(x)**2 + (-2 + np.sin(x)) * np.sin(x) - np.cos(x) * (1 + 4 * np.sin(x)))

def D3f(x):
    return np.exp(np.cos(x) + 2 * np.sin(x)) * (-np.sin(x)**3 + 6 * np.sin(x)**2 + np.sin(x) + 8 * np.cos(x)**3 - 6 * (2 * np.sin(x) + 1) * np.cos(x)**2 + (6 * np.sin(x)**2 - 9 * np.sin(x) - 2) * np.cos(x))

# Initial setup
# Number of equidistance points
N = 7
# Array of N equidistant points over [0,2pi)
nodes = np.linspace(0, 2 * np.pi, N + 1)[:-1]

# Make a 2*2 matrix using nodes that paves the way for the next section
XX = np.subtract.outer(nodes, nodes) / 2
XX += np.eye(N)
id = np.arange(N)
ID = np.subtract.outer(id, id)

# Function and Derivatives (up to three) evaluations 
F = f(nodes)
DF = Df(nodes)
DDF = D2f(nodes)
DDDF = D3f(nodes)

# Differences matrix to satisfy hermite interpolation
# Compute D_i,j to be used in computing g function
j = 0 # For j=0
D1_0 = (j+1) * (-1.)**((j+1) * ID) / 2 * cst(XX, N)
D1_0 = D1_0 - np.diag(np.diag(D1_0)) # Change the diagonal to zero (7*7)
D1_0 = D1_0 - np.diag(np.sum(D1_0, axis=1)) # Change diagonal to the sum of columns (7*7)
DR = D1_0.dot(F) # Apply inner product (1*7)
j = 1 # For j = 1
D2_1 = (j+1) * (-1.)**((j+1) * ID) / 2 * cst(XX, N)
D2_1 = D2_1 - np.diag(np.diag(D2_1))
D2_1 = D2_1 - np.diag(np.sum(D2_1, axis=1))
DDR = D2_1.dot(DF - DR)
D2R = np.linalg.matrix_power(D1_0, 2).dot(F)
# For j = 2
j = 2
D3_2 = (j+1) * (-1.)**((j+1) * ID) / 2 * cst(XX, N) #(7*7)
D3_2 = D3_2 - np.diag(np.diag(D3_2))  # Change the diagonal to zero (7*7)
D3_2 = D3_2 - np.diag(np.sum(D3_2, axis=1)) # Change diagonal to the sum of columns (7*7)
DDDR = D3_2.dot(DDF - DDR) # Apply inner product (1*7)
DD2R = np.linalg.matrix_power(D2_1, 2).dot(DF - DR) # (1*7)
D3R = np.linalg.matrix_power(D1_0, 3).dot(F) #(1*7)

# function d that mentioned in the paper and satisfied hermite conditions
def d(y,i):
    return 2 * np.sin((y - nodes[i]) / 2)
    
# Barycentric form of Berrut's trigonometric function  
def b(y,i):
    return (-1) ** (i) * cst((y - nodes[i]) / 2, N) / np.sum((-1) ** id * cst((y - nodes) / 2, N))
# Interpolation loop
x_eval = np.linspace(0, 2 * np.pi, 1500)
Int = np.zeros_like(x_eval)

for j, y in enumerate(x_eval): # index and points needed to make a graph
    Int[j] = 0 # Iinitial interpolated value that is zero for each index
    for s in range(4): # 3 is the order of derivative that satisfy hermite condition
        for l in range(N): # 7 equidistant nodes
            if s==0:
                g = F # compute g function interduced in theorem 1
            elif s==1:
                g = DF - DR
            elif s==2:
                g = DDF - D2R - DDR
            elif s==3:
                g = DDDF - DDDR - DD2R - D3R
            # Interpolated value
            Int[j] += (d(y, l) ** s * b(y, l) ** (s + 1) * g[l]) / factorial(s)
            
li = np.isin(x_eval, nodes)
liN = np.isin(nodes, x_eval[li])
Int[li] = f(nodes[liN])

MinG = min(min(f(x_eval)), min(Int))
MaxG = max(max(f(x_eval)), max(Int))

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_eval, f(x_eval), '--r', label="Test Function 1")
plt.plot(x_eval, Int, 'b', label="Interpolation")
plt.plot(nodes, f(nodes), '.g', markersize=15, label="Nodes")
plt.axis([0, 2*np.pi, MinG - 1, MaxG + 1])

# Show plot
plt.grid()
plt.title("Compare test function 1 with interpolant satisfied Hermite conditions up to the third derivatives at 7 equidistant nodes")
plt.legend()
plt.show()
```

```python
# Calculate function and derivative of the first test function
def f(x):
    return np.exp(2 * np.sin(x) + np.cos(x))

def Df(x):
    return np.exp(2 * np.sin(x) + np.cos(x)) * (2 * np.cos(x) - np.sin(x))

# Initialize N values
N_values = [5, 10, 20, 40]
err = []

for N in N_values:
    nodes = np.linspace(0, 2 * np.pi, N + 1)[:-1]
    
    
    XX = np.subtract.outer(nodes, nodes) / 2
    XX += np.eye(N)
    id = np.arange(N)
    ID = np.subtract.outer(id, id)

    # Function and derivative evaluations
    F = f(nodes)
    DF = Df(nodes)
    
    j = 0
    D1_0 = (j+1) * (-1.)**((j+1) * ID) / 2 * cst(XX, N)
    D1_0 = D1_0 - np.diag(np.diag(D1_0)) # Change the diagonal to zero (7*7)
    D1_0 = D1_0 - np.diag(np.sum(D1_0, axis=1)) # Change diagonal to the sum of columns (7*7)
    DR = D1_0.dot(F) # Apply inner product (1*7)

    d = lambda y, i: 2 * np.sin((y - nodes[i]) / 2)
    b = lambda y, i: (-1) ** (i) * cst((y - nodes[i]) / 2, N) / np.sum((-1) ** id * cst((y - nodes) / 2, N))

    x_eval = np.linspace(0, 2 * np.pi, 1500)
    Int = np.zeros_like(x_eval)

    for j, y in enumerate(x_eval):
        Int[j]=0
        for s in range(2):  # Order derivatives
            for l in range(N):
                if s==0:
                    g = F
                else:
                    g = DF - DR
                    
                Int[j] += (d(y, l) ** s * b(y, l) ** (s + 1) * g[l]) / factorial(s)
                
    li = np.isin(x_eval, nodes)
    liN = np.isin(nodes, x_eval[li])
    Int[li] = f(nodes[liN])

    err.append(np.max(np.abs(f(x_eval) - Int)))

# Plotting
plt.figure(figsize=(10, 5))
plt.semilogy(N_values[:], err)
plt.title('Absolute Error for the approximation of the first function with N = 5,10,20 and 40')
plt.xlabel('N')
plt.ylabel('Absolute Error')
plt.yscale('log')
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import factorial

def cst(phi, n):
    if n % 2 == 0:
        return 1. / np.tan(phi)
    else:
        return 1. / np.sin(phi)

def Sketch(S, D):
    # Load data based on input D
    if D == 1:
        T = pd.read_excel('Doggo.xls').values
    elif D == 2:
        T = pd.read_excel('Elephant.xls').values
    elif D == 3:
        T = pd.read_excel('ScroogeMcDuck.xls').values
    
    N1 = len(T)
    T1 = np.vstack([T, T[0]])

    # Calculate velocity
    V = (T1[1:N1+1] - T1[:N1]) / (2 * np.pi / (N1 + 1))

    # Subsample
    T = T[::S]
    V = V[::S]
    N = len(T)

    nodes = np.linspace(0, 2*np.pi, N+1)[:-1]
    XX = np.subtract.outer(nodes, nodes) / 2
    XX += np.eye(N)
    id = np.arange(N)
    ID = np.subtract.outer(id, id)

    x_eval = np.linspace(0, 2 * np.pi, 800)

    def d(y, i):
        return 2 * np.sin((y - nodes[i]) / 2)

    def b(y, i):
        return (-1)**(i) * cst((y - nodes[i]) / 2, N) / np.sum((-1)**id * cst((y - nodes) / 2, N))

    # Compute Hermite Interpolation
    Ix = np.zeros_like(x_eval)
    Iy = np.zeros_like(x_eval)
    for idx, y in enumerate(x_eval):
        for s in range(2):  # Order of derivatives
            for l in range(N):
                g_x = T[:, 0] if s == 0 else V[:, 0]
                g_y = T[:, 1] if s == 0 else V[:, 1]
                Ix[idx] += d(y, l)**s * b(y, l)**(s+1) * g_x[l] / factorial(s)
                Iy[idx] += d(y, l)**s * b(y, l)**(s+1) * g_y[l] / factorial(s)

    # Update at nodes
    for i, val in enumerate(x_eval):
        if val in nodes:
            idx = np.where(nodes == val)[0]
            Ix[i] = T[idx, 0]
            Iy[i] = T[idx, 1]

    # Plotting
    plt.figure(figsize=(15, 20))
    plt.subplot(2,1,1)
    plt.scatter(T[:, 0], T[:, 1], color='red', s=2, label='Original Nodes')
    plt.legend()
    plt.title("Nodes")
    plt.subplot(2,1,2)
    plt.plot(Ix, Iy, 'b', label='Interpolated Curve')
    plt.scatter(T[:, 0], T[:, 1], color='red', s=2, label='Original Nodes')
    plt.legend()
    plt.title("Interpolation")
    plt.subplots_adjust(wspace=0.3, hspace=0.3)  # Adjust horizontal space between plots
    plt.show()    
# Apply interpolation with all nodes
Sketch(1, 1)
```
